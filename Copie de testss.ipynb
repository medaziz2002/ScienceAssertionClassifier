```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from scipy.sparse import hstack

# Fonction de nettoyage am√©lior√©e
def clean_data(data, column_name):
    data[column_name] = data[column_name].astype(str)
    data[column_name] = data[column_name].apply(lambda text: text.lower())
    return data

# Fonction pour enlever les URLs
def remove_urls(text):
    return re.sub(r"http[s]?://\S+", "", text)

# Fonction pour enlever les caract√®res sp√©ciaux (sauf @ et # si tu veux garder mentions/hashtags dans le texte)
def remove_special_characters(text):
    return re.sub(r"[^a-zA-Z0-9\s@#]", "", text)

# Fonction pour enrichir les donn√©es avec les nouvelles features
def enrich_data(data, column_name):
    data["tweet_length"] = data[column_name].apply(len)
    data["nb_urls"] = data[column_name].apply(lambda x: len(re.findall(r'http[s]?://\S+', x)))
    data["nb_mentions"] = data[column_name].apply(lambda x: x.count('@'))
    data["nb_hashtags"] = data[column_name].apply(lambda x: x.count('#'))
    return data

# Fonction globale de pr√©paration
def prepare_data(data, column_name="text"):
    # Nettoyage de base
    data = enrich_data(data, column_name)  # d'abord enrichir car sinon URLs sont supprim√©es
    data = clean_data(data, column_name)   # nettoyage
    return data

# Exemple d'utilisation
dataClean = prepare_data(dataClean, column_name="text")

# TF-IDF
vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=10000)
X_text = vectorizer.fit_transform(dataClean["text"])

# Concatenation TF-IDF + nouvelles features
X_additional = dataClean[["tweet_length", "nb_urls", "nb_mentions", "nb_hashtags"]]
X_final = hstack([X_text, X_additional])

# Cible
y = dataClean["science_related"]

# SMOTE pour √©quilibrer
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_final, y)

# 1. Nettoyer le texte brut
dataClean = enrich_data(data, column_name="text")  # Ajoute longueur, urls, mentions, hashtags
dataClean = clean_data(dataClean, column_name="text")   # Nettoyage de la colonne text

# 2. Vectoriser apr√®s nettoyage
vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=10000)
X_text = vectorizer.fit_transform(dataClean["text"])  # ICI, dataClean["text"] est un texte propre

# 3. Ajouter les autres features (longueur, urls, hashtags, mentions)
X_additional = dataClean[["tweet_length", "nb_urls", "nb_mentions", "nb_hashtags"]]

# 4. Fusionner TF-IDF et features
from scipy.sparse import hstack
X_final = hstack([X_text, X_additional])

# 5. Cible
y = dataClean["science_related"]

# 6. SMOTE pour √©quilibrer
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_final, y)

# 7. Run classification
run_classification(X=X_res, y=y_res, model_type="random_forest", n_classes=2)

```
