#1- Fonction qui remplace les emoji par des chaÃ®nes de caractÃ¨res

```python
def replace_emojis(text):
    return emoji.demojize(text, delimiters=(" ", " "))
```

#2- Fonction qui supprime les URLS

```python
def remove_urls(text):
    """Remove URLs from text."""
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```

#3- Fonction qui retire les mentions

```python
def remove_mentions(text):
    """Remove @mentions from text."""
    return re.sub(r'@\w+', '', text)
```

#4- Fonction qui retire les hashtags

```python
def remove_hashtags(text):
    """Remove #hashtags from text."""
    return re.sub(r'#\w+', '', text)
```

#5- Fonction qui retire la ponctuation

```python
def remove_punctuation(text):
    """Remove punctuation from text."""
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

#6- Fonction qui retire les nombres

```python
def remove_numbers(text):
    """Remove numbers from text."""
    return re.sub(r'\d+', '', text)
```

#7- Fonction qui retire les stopwords

```python
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')



def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    """
    Supprime les stop words d'un texte.

    :param text: str - Le texte d'entrÃ©e (un tweet dans ce cas).
    :return: str - Le texte nettoyÃ© sans stop words.
    """
    words = text.split()  # Tokenisation simple
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

#9- Fonction total

```python
def preprocess_text(text):
    """Apply all preprocessing steps to text."""
    if pd.isna(text) or text == '':
        return ''

    text = str(text).lower()
    text = replace_emojis(text)
    text = remove_urls(text)
    text = remove_urls(text)
    text = remove_mentions(text)
    text = remove_hashtags(text)
    text = remove_punctuation(text)
    text = remove_numbers(text)
    text = remove_stopwords(text)
    text = lemmatize_text(text)
    return text
```

```python
def preprocess_dataset(df, text_column='text'):
    """Apply preprocessing to the entire dataset."""
    # Create a copy to avoid modifying the original
    df_processed = df.copy()

    # Apply text preprocessing to the specified column
    df_processed['text_processed'] = df_processed[text_column].apply(preprocess_text)

    return df_processed
```
