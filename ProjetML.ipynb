```python

import pandas as pd
import emoji

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ProjetML/data/scitweets_export.tsv', sep='\t')

display(data.head())
```

#1- Ingénierie des données

##1.1- Fonction qui remplace les emoji par des chaînes de caractères

```python
def replace_emojis(text):
    return emoji.demojize(text, delimiters=(" ", " "))
```

```python
data['text'] = data['text'].apply(replace_emojis)
display(data.head())
```

##1.2- Fonction qui retire les ponctuations et les hashtags

```python
import re
def clean_punct(text):
    #retirer les hashtags
    text = re.sub(r'#\w+', '', text)
    #retirer les mentions
    text = re.sub(r'@\w+', '', text)
    #retirer la ponctuation
    text = re.sub(r'[^\w\s]', '', text)
    #retirer les espaces
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
data['text'] = data['text'].apply(clean_punct)
print(data[['text']])
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def lemmatize_text(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])
```

```python
data['text'] = data['text'].apply(lemmatize_text)
display(data.head())
```

##1.4- Fonction qui retire les StopWords

```python
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    """
    Supprime les stop words d'un texte.

    :param text: str - Le texte d'entrée (un tweet dans ce cas).
    :return: str - Le texte nettoyé sans stop words.
    """
    words = text.split()  # Tokenisation simple
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

Nouvelle variable dataNoStopwords qui représente toutes les données sans stopwords\

```python
data['text'] = data['text'].apply(remove_stopwords)
display(data.head())
```
